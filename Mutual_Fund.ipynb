```python
import pandas as pd
import numpy as np

df = pd.read_csv("/content/Mutual Fund Dataset 2025.csv")

print(df.shape)
df.head()
```

```python
# Replacing missing symbols with NaN
df.columns = df.columns.str.strip()
df.replace(['-', ' - ', '--', 'NA', 'N/A'], np.nan, inplace=True)
```

```python
for col in df.select_dtypes(include='object').columns:
    df[col] = df[col].str.strip()
```

```python
percentage_columns = [
    'ExpenseRatio (%)',
    'Return (%)1 mo',
    'Return (%)3 mo',
    'Return (%)6 mo',
    'Return (%)1 yr',
    'Return (%)2 yrs',
    'Return (%)3 yrs',
    'Return (%)5 yrs',
    'Return (%)10 yrs',
    'Turnover Ratio (%)',
    'Large Cap(%)',
    'Mid Cap(%)',
    'Small Cap(%)',
    'Yield To Maturity (%)'
]

for col in percentage_columns:
    if col in df.columns:
        df[col] = df[col].str.replace('%', '', regex=False)
        df[col] = pd.to_numeric(df[col], errors='coerce')
```

```python
# Convert Inception Date to Datetime

df['Inception Date'] = pd.to_datetime(df['Inception Date'],errors='coerce')
```

```python
# Convert Float Columns

float_columns = [
    'AUM(in Rs. cr)', 'ExpenseRatio (%)', 'NAV',
    '52 WeekHigh (NAV)', '52 WeekLow (NAV)',
    'Return (%)1 mo', 'Return (%)3 mo', 'Return (%)6 mo',
    'Return (%)1 yr', 'Return (%)2 yrs', 'Return (%)3 yrs',
    'Return (%)5 yrs', 'Return (%)10 yrs',
    'Turnover Ratio (%)', 'Avg. Market Cap(in Rs. cr)',
    'Large Cap(%)', 'Mid Cap(%)', 'Small Cap(%)',
    'Avg. Maturity(in yrs)', 'Mod. Duration(in yrs)',
    'Yield To Maturity (%)', 'Alpha', 'Sharpe',
    'Sortino', 'Beta', 'Standard Deviation'
]

for col in float_columns:
    if col in df.columns:
        df[col] = pd.to_numeric(df[col], errors='coerce')
```

```python
print(df.columns.tolist())
```

```python
# Convert Integer Column

df['No. ofStocks'] = pd.to_numeric(df['No. ofStocks'], errors='coerce')
df['No. ofStocks'] = df['No. ofStocks'].fillna(0).astype(int)
```

```python
# Handle Null Values

for col in float_columns + ['No. ofStocks']:
    if col in df.columns:
        df[col] = df[col].fillna(df[col].median())

categorical_columns = [
    'Funds',
    'Fund Manager',
    'Category',
    'RupeeVestRating',
    'Fund Type',
    'Benchmark Index',
    'Highest Sector',
    'classification',
    'Exit_load_Remarks'
]

for col in categorical_columns:
    if col in df.columns:
        df[col] = df[col].fillna("Unknown")
```

```python
# Remove Duplicates

df.drop_duplicates(inplace=True)
```

```python
print(df.dtypes)
df.shape
```

```python
# Save Cleaned Dataset

df.to_csv("cleaned_mutualfund_dataset.csv", index=False)
```
